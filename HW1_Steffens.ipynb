### (a) Find the size of the data frame. How many rows, how many columns?

```python
rows, cols = medals_df.shape
print(f"The DataFrame has {rows} rows and {cols} columns.")
```

### (b) Use the appropriate function to report the data types for all columns.

```python
print("Data types for each column:")
print(medals_df.dtypes)
```

### (c) Find how many unique cities matches were held at.

```python
num_unique_cities = medals_df['Location'].nunique()
print(f"Matches were held at {num_unique_cities} unique cities.")
```

### (d) Find how many medals the USA team won in total.

```python
usa_medals = medals_df[medals_df['Nationality'] == 'USA']
total_usa_medals = len(usa_medals)
print(f"The USA team won a total of {total_usa_medals} medals.")
```

### (e) Find the total number of medals for each Nationality.

```python
medals_by_nationality = medals_df.groupby('Nationality').size().sort_values(ascending=False)
print("Total number of medals for each Nationality:")
display(medals_by_nationality)
```
