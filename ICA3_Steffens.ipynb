```python
import csv

Price = {}
Difficulty = {}

file_path = '/content/drive/MyDrive/DATA/lego_setsHB.csv'


with open(file_path, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    header = next(reader)

    product_id_idx = header.index('prod_id')
    price_idx = header.index('list_price')
    difficulty_idx = header.index('review_difficulty')

    for i, row in enumerate(reader):
        if i >= 1000:
            break

        if len(row) > max(product_id_idx, price_idx, difficulty_idx):
            product_id = row[product_id_idx]
            price = float(row[price_idx])
            difficulty = row[difficulty_idx]

            if product_id:
                Price[product_id] = price
                Difficulty[product_id] = difficulty

print("\nFirst 5 entries of Price dictionary:")
for k, v in list(Price.items())[:5]:
    print(f"  {k}: {v}")

print("\nFirst 5 entries of Difficulty dictionary:")
for k, v in list(Difficulty.items())[:5]:
    print(f"  {k}: {v}")
```
