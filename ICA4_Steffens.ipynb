```python
import pandas as pd

file_path = '/content/drive/MyDrive/lego_sets.csv'
df = pd.read_csv(file_path)

print("DataFrame 'df' loaded successfully. Displaying the first 5 rows:")
df.head()
```

```python
print("DataFrame Info:")
df.info()

print("\nDataFrame Description:")
df.describe()
```

```python
high_rated_sets = df[df['star_rating'] >= 4]

print("Lego sets with a star_rating >= 4. Displaying the first 5 rows:")
high_rated_sets.head()
```

```python
num_high_rated_sets = len(high_rated_sets)

print(f"Number of Lego sets with a star_rating >= 4: {num_high_rated_sets}")
```

```python
average_price_high_rated_sets = high_rated_sets['list_price'].mean()

print(f"Average list price for Lego sets with a star_rating >= 4: ${average_price_high_rated_sets:.2f}")
```
